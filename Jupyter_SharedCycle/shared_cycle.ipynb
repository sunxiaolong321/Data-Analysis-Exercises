# 共享单车使用情况分析

```python
import pandas as pd
from pyecharts.charts import Line,Pie,Bar,Funnel,HeatMap
from pyecharts import options as opts
import os
```

```python
os.listdir(path=r".\data")
```

```python
lis = []
for csv in os.listdir(path=r".\data"):
    lis.append(pd.read_csv(r".\data\\"+csv))
data = pd.concat(lis)
```

```python
data.head()
```

# 插入骑行花费时间列并转成分钟

```python
data.insert(9,"minute",data["Duration (ms)"])
data.astype({"minute":"float"})
data["minute"] = data["minute"].apply(lambda x:round(x/(1000*60),2))
```

# 对数据索引恢复，修改时间格式

## 索引重新排序

```python
data.reset_index(inplace=True)
```

# 修改时间格式

```python
data['Start date'] = pd.to_datetime(data['Start date'])
data['End date'] = pd.to_datetime(data['Start date'])
```

```python
data.info()
```

### 提取出2017年的月份，中心组织表格

```python
data["hour"] = data["Start date"].dt.hour
data["month"] = data["Start date"].dt.month
```

```python
data.head()
```

## 统计一年中平均每小时骑行时长

### 数据分组

```python
hour = data["minute"].groupby([data["hour"],data["Member type"]]).sum()
count = data["minute"].groupby([data["hour"],data["Member type"]]).count()
```

```python
count
```

```python
# 数据扁平化
casual_time = hour.loc[(slice(None),'Casual')]
member_time = hour.loc[(slice(None),'Member')]
casual_count = count.loc[(slice(None),'Casual')]
member_count = count.loc[(slice(None),'Member')]
```

### 数据画图

```python
line = (
    Line()
    .add_xaxis(list(casual_time.index))
    .add_yaxis(
        '非会员骑行时间',
        [round(x,2) for x in casual_time.values],
        is_smooth=True,
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[opts.MarkPointItem(type_="max"),
                  opts.MarkPointItem(type_="min")],
        )
    )
#     .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}时: {c}"))
    .add_yaxis(
        '会员骑行时间',
        [round(x,2) for x in member_time.values],
        is_smooth=True,
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[opts.MarkPointItem(type_="max"),
                  opts.MarkPointItem(type_="min")],
        )
    )
    .add_yaxis(
        '非会员骑行人员数量',
        casual_count.values.tolist(),
        is_smooth=True,
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[opts.MarkPointItem(type_="max"),
                  opts.MarkPointItem(type_="min")],
        )
    )
    .add_yaxis(
        '会员骑行人员数量',
        member_count.values.tolist(),
        is_smooth=True,
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[opts.MarkPointItem(type_="max"),
                  opts.MarkPointItem(type_="min")],
        )
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="共享单车时辰骑行人数"))
)
```

```python
line.render_notebook()
```

## 骑行时长分布

```python
time_series = pd.value_counts(pd.cut(data["minute"].values ,[0,1,3,5,10,15,30,60,90,120,180,300,600,100000])).sort_index()
```

```python
time_series
```

```python
time_division = ["1分钟以内","1分钟~3分钟","3分钟~5分钟","5分钟~10分钟","10分钟~15分钟","15分钟~30分钟","30分钟~60分钟",\
               "60分钟~90分钟","90分钟~120分钟","120分钟~180分钟","180分钟~300分钟","300分钟~600分钟","600分钟以上"]
```

```python
bar = (
    Bar()
    .add_xaxis(time_division)
    .add_yaxis(
        "骑行时间",
        time_series.to_list(),
         category_gap=0
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="整年骑行时长"),
        toolbox_opts = opts.ToolboxOpts(is_show=True)
        )
)
bar.render_notebook()
```

```python
funnel = (
    Funnel()
    .add(
        "整年骑行时长",
        [list(z) for z in zip(time_division,time_series.to_list())],
        sort_="ascending",
        label_opts=opts.LabelOpts(position="inside"),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="漏斗图"))
)
funnel.render_notebook()
```

## 骑行时长和月份之间的关系

```python
time_series = pd.value_counts(pd.cut(data["minute"].values ,[0,1,3,5,10,15,30,60,90,120,180,300,600,100000]))
```

```python
liss = []
for x,y in data["minute"].groupby(data["month"]):
    count = 0
    for z in pd.value_counts(pd.cut(y.values ,[0,1,3,5,10,15,30,60,90,120,180,300,600,100000])).sort_index():
        lis = []
        lis.append(x-1)
        lis.append(count)
        lis.append(z)
        count += 1
        liss.append(lis)
```

```python
heatmap = (
    HeatMap()
    .add_xaxis(["%s月"%i for i in range(1,12)])
    .add_yaxis("series0", time_division, liss)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="骑行时长热力图"),
        visualmap_opts=opts.VisualMapOpts(max_=15000),
    )
)
```

```python
heatmap.render_notebook()
```
